```python
import numpy as np
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
maximo_palavras = 10000
(x_treino, y_treino), (x_teste, y_teste) = imdb.load_data(num_words=maximo_palavras)
```

```python
maximo_comprimento = 200
# o pad vai aumentar textos com menos de 200 palavras e cortar textos com mais de 200
x_treino = pad_sequences(x_treino, maxlen=maximo_comprimento)
x_teste = pad_sequences(x_teste, maxlen=maximo_comprimento)
```

```python
# Agora vamos usar a função e para vetorizar as palavras em números (embedding)
embedding_dim = 128
modelo = Sequential()
modelo.add(Embedding(maximo_palavras, embedding_dim, input_length=maximo_comprimento))
```

```python
# Aqui processamos os embeddings e mantemos o contexto
modelo.add(LSTM(128))
```

```python
# Agora forçaremos os outputs a serem 0 ou 1 (positivo ou negativo) (Sigmoid)
modelo.add(Dense(1, activation='sigmoid'))
```

```python
# Fazemos a compilação para definir como a máquina aprende e quais resultados queremos minimizar
modelo.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
modelo.summary()
```

```python
# Treinamento do Modelo
historico = modelo.fit(x_treino, y_treino, epochs=10, batch_size=32,validation_split=0.2)
# epoch é o numero de vezes que o modelo verá o conjuto de dados INTEIRO
# batch size signfica quantas extrapolações ele vai fazer antes de atualizar pesos
```

```python
loss, accuracy = modelo.evaluate(x_teste, y_teste, verbose=0)
print(f'Precisão no conjunto de teste: {accuracy*100:.2f}%')
```
